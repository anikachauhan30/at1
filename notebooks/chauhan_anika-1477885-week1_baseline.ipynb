Data Preprocessing

```python
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
features = df.drop(['player_id','drafted','ftr','pfr',],axis=1)
target = df['drafted']
features = features.dropna()
```

```python
features = pd.get_dummies(features)
```

```python
features.describe()
```

Distribution of Target

```python
sbn.countplot(x='drafted',data=df)
plt.title("Distribution of Drafted players")
plt.show()
```

Correlation between different features

```python
sbn.heatmap(features.corr(),cmap='coolwarm')
plt.title("correlation matrix")
plt.show()
```

Baseline

```python
average_draft_prob = target.mean()
print(f'Average Draft Probability for all players (Baseline): {average_draft_prob:.4f}')
```

```python
# Convert categorical features (e.g., team, conf, yr) into numerical using one-hot encoding
data = pd.get_dummies(df, columns=['team', 'conf','ftr', 'yr', 'pfr', 'year','type','ht','player_id' ])
```

```python
sbn.heatmap(plt.matshow(data.corr(),annot=True,cmap='coolwarm'))
plt.title("Correlation Matrix")
plt.show()
```
